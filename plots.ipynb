# Monte Carlo

```python
n_s = [100]
s_eta_s = [0.99]
lamb_s = [0.5, 1, 2]
lambda_W_s = [0.2, 0.5, 1]
lambda_X_s = [0.2, 0.5, 1]
```

## different lambdas

```python
for n, s_eta, lambda_W, lambda_X in product(n_s, s_eta_s, lambda_W_s, lambda_X_s):
    mc_data = MonteCarloData(
                            n_s=[n],
                            d=50,
                            K=20,
                            s_eta_s=[s_eta],
                            lamb_s=lamb_s,
                            lambda_W_s=[lambda_W],
                            lambda_X_s=[lambda_X],
                            compare="lambda"
                            )
    print(f"n: {n} | s_eta: {s_eta} | lambda_W: {lambda_W} | lambda_X: {lambda_X}")
    mc_data.create_plots(save=True)
```

## different lambda_W s

```python
for n, s_eta, lamb, lambda_X in product(n_s, s_eta_s, lamb_s, lambda_X_s):
    mc_data = MonteCarloData(
                            n_s=[n],
                            d=50,
                            K=20,
                            s_eta_s=[s_eta],
                            lamb_s=[lamb],
                            lambda_W_s=lambda_W_s,
                            lambda_X_s=[lambda_X],
                            compare="lambda_W"
                            )
    print(f"n: {n} | s_eta: {s_eta} | lambda: {lamb} | lambda_X: {lambda_X}")
    mc_data.create_plots(save=True)
```

## different lambda_X s

```python
for n, s_eta, lamb, lambda_W in product(n_s, s_eta_s, lamb_s, lambda_W_s):
    mc_data = MonteCarloData(
                            n_s=[n],
                            d=50,
                            K=20,
                            s_eta_s=[s_eta],
                            lamb_s=[lamb],
                            lambda_W_s=[lambda_W],
                            lambda_X_s=lambda_X_s,
                            compare="lambda_X"
                            )
    print(f"n: {n} | s_eta: {s_eta} | lambda: {lamb} | lambda_W: {lambda_W}")
    mc_data.create_plots(save=True)
```
